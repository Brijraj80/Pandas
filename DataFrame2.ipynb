```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
df = pd.DataFrame(randn(8,8),['A','B','C','D','E','F','G','H'],['USA','India','Canada','China','Japan','Korea','Italy','Thailand'])
```

```python
df
```

```python
#Coditional Selection
#NaN represent false values with doesn't comply with our condition  
df[df>0]
```

```python
df['India']>0
```

```python
#It return values which are true , doesn't retrive false values
df[df['USA']>0]
```

```python
df[df['USA']<0]
```

```python
df.iloc[[1,2],[0,1]]
```

```python
df[df['USA']>0][['USA','India','China']]
```

```python
#you can't use normal "and" while dealing with multiple condition as it can hold only one boolean value T and F = F
#Use & instead
df[(df['USA']>0) & (df['India']>0) & (df['China']>0)][['USA','India','China']]
```

```python
#you can't use normal "or" while dealing with multiple condition as it can hold only one boolean value T and F = F
#Use '|' instead
df[(df['USA']>0) | (df['India']>0)][['USA','India']]
```
