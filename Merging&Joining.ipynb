```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['Abascus','Astonia','Altrax','Astro','Azer'],
                    'B':['Branford','Bogus','Brandon','Baaccl','Boston'],
                    'C':['Caucus','Cataline','Cathrine','Cacta','Costa'],
                    'D':['Dominar','Democrats','Demolish','Dasda','Drek']},index=[1,2,3,4,5])
```

```python
df1
```

```python
df2 = pd.DataFrame({'A':['A1','A2','A3','A4','A5'],
                    'B':['B1','B2','B3','B4','B5'],
                    'C':['C1','C2','C3','C4','C5'],
                    'D':['D1','D2','D3','D4','D5']},index=[6,7,8,9,10])
```

```python
df2
```

```python
df3 = pd.DataFrame({'A':['K1','K2','K3','K4',np.nan],
                    'B':['L1','L2','L3','L4',np.nan],
                    'C':['M1','M2','M3','M4','M5'],
                    'D':['N1','N2','N3','N4',np.nan]},index=[11,12,13,14,15])
```

```python
df3
```

```python
#Column must be same for all DataFrames
#This concat the rows
pd.concat([df1,df2,df3])
```

```python
#This concat the columns 
pd.concat([df1,df2,df3],axis=1)
```

```python
#Merging the DataFrames
```

```python
left = pd.DataFrame({'A':['A1','A2'],'B':['B1','B2'],'C':['C1','C2'],'Join':['D1','D2']})
```

```python
right = pd.DataFrame({'E':['E1','E2'],'F':['F1','F2'],'Join':['D1','D2']})
```

```python
left 
```

```python
right
```

```python
#Values of 'Join' must be same in both the table , similar to join operation in SQL
pd.merge(left,right,how='inner',on='Join')
```

```python
left = pd.DataFrame({'A':['A1','A2'],'B':['B1','B2'],'C':['C1','C2']},index=[1,2])
```

```python
right = pd.DataFrame({'E':['E1','E2','E3'],'F':['F1','F2','F3']},index=[1,2,3])
```

```python
left.join(right)
```
